```python
import random
import collections

class MarkovChain:
    def __init__(self):
        self.transitions = collections.defaultdict(list)

    def train(self, text, order=1):
        words = text.split()
        for i in range(len(words) - order):
            state = tuple(words[i:i + order])
            next_word = words[i + order]
            self.transitions[state].append(next_word)

    def generate(self, length=50, seed=None):
        if seed is None:

            if not self.transitions:
                return ""
            seed = random.choice(list(self.transitions.keys()))
        current_state = seed
        result = list(current_state)
        for _ in range(length - len(current_state)):

            if not self.transitions[current_state]:
                break
            next_word = random.choice(self.transitions[current_state])
            result.append(next_word)
            current_state = tuple(result[-len(current_state):])
        return ' '.join(result)
```

```python
text = """It was the best of times, it was the worst of times, it was the age of wisdom, it was the age of foolishness,
it was the epoch of belief, it was the epoch of incredulity, it was the season of Light, it was the season of Darkness,
it was the spring of hope, it was the winter of despair, we had everything before us, we had nothing before us,
we were all going direct to Heaven, we were all going direct the other way â€“ in short, the period was so far like the present period,
that some of its noisiest authorities insisted on its being received, for good or for evil, in the superlative degree of comparison only."""
```

```python
chain = MarkovChain()
chain.train(text, order=2)
generated_text = chain.generate(length=100)
print(generated_text)
```
